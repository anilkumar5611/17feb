```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios is it preferred to use MongoDB over SQL databases?

MongoDB: MongoDB is a popular NoSQL, document-oriented database that provides high performance, high availability, and easy scalability. It stores data in flexible, JSON-like documents called BSON (Binary JSON), allowing for the representation of complex hierarchical relationships.

Non-relational databases: Non-relational databases, also known as NoSQL databases, are designed to handle unstructured, semi-structured, or structured data. Unlike traditional relational databases (SQL databases), NoSQL databases don't rely on a fixed schema, making them more flexible and scalable. They are suitable for scenarios where data is constantly evolving, and the volume of data is large and distributed.

Scenarios to use MongoDB over SQL databases:

When dealing with unstructured or semi-structured data.
In scenarios where the schema may evolve frequently.
For horizontal scaling and distributing data across multiple servers.
Rapid development and prototyping, as MongoDB is schema-less.
When flexible data models are required.
```

```python
Q2. State and Explain the features of MongoDB.

Features of MongoDB:

Document-Oriented: Stores data in flexible, JSON-like BSON documents, allowing for nested structures and arrays.

Schema-less: No fixed schema, making it adaptable to changing data requirements.

Query Language: Supports rich query expressions, including field, range, and regular expression searches.

Indexes: Allows the creation of indexes to improve query performance.

Aggregation Framework: Provides powerful and flexible aggregation capabilities for data processing.

Replication: Supports automatic and manual replication for high availability and fault tolerance.

Sharding: Enables horizontal scaling by distributing data across multiple servers.

Geospatial Indexes: Supports geospatial indexing for location-based queries.

GridFS: Allows storage and retrieval of large files, exceeding the BSON document size limit.

Security: Provides authentication, authorization, and SSL/TLS encryption for secure data storage.

Atomic Operations: Supports atomic operations on single documents, providing consistency.

```

```python
Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.

python
Copy code
# Install pymongo using: pip install pymongo
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create or access a database (replace 'mydatabase' with your preferred database name)
mydb = client["mydatabase"]

# Create or access a collection (replace 'mycollection' with your preferred collection name)
mycollection = mydb["mycollection"]

print("Connected to MongoDB.")
```

```python
Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

Using the find() method:
The find() method is used to retrieve documents from a collection in MongoDB. It returns a cursor that can be iterated to access the documents.

python
Copy code
# Querying with find() method
query = {"city": "New York"}
result = mycollection.find(query)

# Print the matching documents
print("Matching Documents:")
for document in result:
    print(document)
This code queries for documents where the "city" field is "New York" and prints the matching documents.

```

```python
Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

sort() method:
The sort() method in MongoDB is used to sort the result of a query in ascending (1) or descending (-1) order based on one or more fields.

Example:

python
Copy code
# Sorting documents by age in descending order
sorted_result = mycollection.find().sort("age", -1)

# Print the sorted documents
print("Sorted Documents by Age (Descending):")
for document in sorted_result:
    print(document)
In this example, the documents are sorted based on the "age" field in descending order (-1).

```
